```python
import math                              # Funciones matematicas
import matplotlib.pyplot as plt          # Para graficar

# METODO DE NEWTON CLASICO Y DE LA CUERDA


# PARTE A. COMPLETAR EN CLASE
# Aqui se define la funcion y su derivada


def f(x):
    # Completar en clase:
    # Ejemplo 1:
    return math.cos(2*x)**2 - x**2
    


def df(x):
    # Completar en clase:
    # Derivada del Ejemplo 1:
    return -4*math.cos(2*x)*math.sin(2*x) - 2*x  
   



# PARTE B. COMPLETAR EN CLASE
# Datos iniciales


x0 = 0.75                 # Punto inicial
tol = 1e-10               # Tolerancia
max_iter = 100            # Maximo numero de iteraciones



# PARTE C. BLOQUE BASE: NEWTON CLASICO


def newton_clasico(x0, tol=1e-10, max_iter=100):
    x = x0
    errores = []                                 #Lista para guardar errores
    iteraciones = []                             #Lista para guardar las iteraciones x_k

    print("\nMETODO DE NEWTON CLASICO")
    print("Iter |      x_k       |     f(x_k)      |     f'(x_k)     | Error absoluto")
    print("-------------------------------------------------------------------------------")

    for k in range(max_iter):                      
        fx = f(x)                                 #Evalua la función en x_k
        dfx = df(x)                               #Evalúa f' en x_k

        if dfx == 0:                              #Evita la división por cero
            print("La derivada se hizo cero. El metodo se detiene.")
            break

        # FORMULA DE NEWTON CLASICO
        x_new = x - fx / dfx                    

        error = abs(x_new - x)                  # Error absoluto entre iteraciones

        print(f"{k:4d} | {x:14.10f} | {fx:14.10e} | {dfx:14.10e} | {error:14.10e}")

        errores.append(error)                  # guardar el error
        iteraciones.append(x)                  # guardar x_k

        if error < tol:                        #criterio de parada
            x = x_new                          #actualiza antes de salir
            iteraciones.append(x)              # guarda la última aproximación
            break

        x = x_new                              #actualiza x para la siguiente iteración

    print("\nRespuesta final Newton clasico:")
    print(f"Raiz aproximada = {x:.12f}")
    print(f"Numero de iteraciones = {k+1}")

    return x, errores, iteraciones



# PARTE D. BLOQUE BASE: MÉTODO DE LA CUERDA


def newton_modificado(x0, tol=1e-10, max_iter=100):
    x = x0
    errores = []
    iteraciones = []

    # DERIVADA FIJA EN x0
    dfx0 = df(x0)

    print("\nMETODO DE NEWTON MODIFICADO")
    print("Iter |      x_k       |     f(x_k)      |   f'(x0) fija    | Error absoluto")
    print("-------------------------------------------------------------------------------")

    if dfx0 == 0:
        print("La derivada inicial es cero. El metodo no puede comenzar.")
        return x, errores, iteraciones

    for k in range(max_iter):
        fx = f(x)                             #Evaluar la función en x_k

        # FORMULA DE MÉTODO DE LA CUERDA
        x_new = x - fx / dfx0   

        error = abs(x_new - x)                #error absoluto de las iteraciones

        print(f"{k:4d} | {x:14.10f} | {fx:14.10e} | {dfx0:14.10e} | {error:14.10e}")

        errores.append(error)                  # guarda el error
        iteraciones.append(x)                  #guarda las iteraciones

        if error < tol:                         
            x = x_new
            iteraciones.append(x)
            break

        x = x_new

    print("\nRespuesta final Newton modificado:")
    print(f"Raiz aproximada = {x:.12f}")
    print(f"Numero de iteraciones = {k+1}")

    return x, errores, iteraciones



# PARTE E. EJECUCION


raiz_newton, err_newton, xs_newton = newton_clasico(x0, tol, max_iter)              # Ejecuta el método de newton clásico
raiz_mod, err_mod, xs_mod = newton_modificado(x0, tol, max_iter)                    # Ejecuta el método de newton mdificado



# PARTE F. GRAFICA


plt.figure(figsize=(8, 5))
plt.semilogy(range(1, len(err_newton)+1), err_newton, marker='o', label='Newton clasico')
plt.semilogy(range(1, len(err_mod)+1), err_mod, marker='s', label='Método de la Cuerda')
plt.xlabel('Iteracion')
plt.ylabel('Error absoluto')
plt.title('Comparacion de convergencia')
plt.grid(True)
plt.legend()
plt.show()
```

# Comparación: Método de Newton vs Método de la Cuerda

| Aspecto | Newton clásico | Newton modificado |
|--------|--------------|------------------|
| Fórmula | $ x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)} $ | $ x_{k+1} = x_k - \frac{f(x_k)}{f'(x_0)} $ |
| Uso de la derivada | Se actualiza en cada iteración | Se calcula una sola vez |
| Velocidad de convergencia | 🚀 Cuadrática (muy rápida) | 🐢 Lineal (más lenta) |
| Número de iteraciones | Pocas | Muchas |
| Costo por iteración | Alto (derivada cada vez) | Bajo (derivada fija) |
| Precisión | Muy alta en pocas iteraciones | Requiere más iteraciones para igual precisión |
| Estabilidad | Puede fallar si \( f'(x_k) \approx 0 \) | Más estable si la derivada no cambia mucho |
| Implementación | Más costosa | Más simple |
| Uso típico | Problemas pequeños o académicos | Problemas grandes o costosos |
| Ejemplo en gráfica | Caída abrupta del error | Disminución gradual del error |

---

## Ventajas

### Newton clásico
- Convergencia muy rápida (cuadrática)
- Alta precisión en pocas iteraciones
- Método estándar en teoría

### Método de la Cuerda
- Menor costo computacional por iteración
- Más fácil de implementar en problemas grandes
- Útil cuando la derivada es costosa

---

## Desventajas

### Newton clásico
- Requiere calcular la derivada en cada iteración
- Puede ser costoso en problemas grandes
- Sensible a malas aproximaciones iniciales

### Método de la Cuerda
- Convergencia más lenta
- Puede requerir muchas iteraciones
- Menor eficiencia global en problemas pequeños
