# 📘 Resumen de Teoría: Métodos de Newton

## 🔵 1. Método de Newton estándar (una variable)

Se usa para resolver:
$$
f(x)=0
$$

### Iteración:
$$
x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)}
$$

### Idea:
- Se aproxima la función por su recta tangente.
- La intersección con el eje $x$ da la siguiente aproximación.

### Propiedades:
- Convergencia **cuadrática** si la raíz es simple.
- Requiere evaluar la derivada en cada iteración.
- Sensible al punto inicial.

### Criterio de parada:
$$
|x_{k+1} - x_k| < \varepsilon
$$

---

## 🔴 2. Raíces múltiples

Si la raíz tiene multiplicidad $m > 1$:

👉 Newton estándar pierde rapidez (convergencia lineal).

---

## 🟢 3. Newton modificado (raíces múltiples)

Si se conoce la multiplicidad $m$:

### Iteración:
$$
x_{k+1} = x_k - m \frac{f(x_k)}{f'(x_k)}
$$

### Ventaja:
- Recupera convergencia rápida.

### Desventaja:
- Se necesita conocer $m$.

---

## 🟡 4. Newton adaptativo

Cuando no se conoce $m$:

### Iteración:
$$
x_{k+1} = x_k - m^{(k)} \frac{f(x_k)}{f'(x_k)}
$$

### Idea:
- Estima la multiplicidad durante las iteraciones.

### Ventaja:
- No requiere conocer $m$.
- Mejora convergencia en raíces múltiples.


## 🔵 Estimación de la multiplicidad en el método de Newton adaptativo

En el método de Newton adaptativo, la multiplicidad de la raíz no se conoce previamente.  
Por ello, se estima en cada iteración mediante una cantidad $m^{(k)}$.

---

### 📌 Fórmula de estimación

La multiplicidad aproximada en la iteración $k$ se calcula como:

$$
m^{(k)} = \frac{x^{(k-1)} - x^{(k-2)}}{2x^{(k-1)} - x^{(k)} - x^{(k-2)}}
$$

---

### 🧠 Interpretación

- Se utilizan **tres iteraciones consecutivas**:
  - $x^{(k)}$
  - $x^{(k-1)}$
  - $x^{(k-2)}$

- La fórmula analiza cómo cambian las aproximaciones para estimar la multiplicidad.

---

### 🔧 Uso en el método

Una vez calculado $m^{(k)}$, se usa en la iteración de Newton:

$$
x_{k+1} = x_k - m^{(k)} \frac{f(x_k)}{f'(x_k)}
$$

---

### 🧩 Procedimiento

1. Realizar al menos **dos iteraciones de Newton estándar**:
   - obtener $x^{(0)}, x^{(1)}, x^{(2)}$

2. Calcular $m^{(2)}$ con la fórmula

3. A partir de ahí, aplicar el método adaptativo en cada iteración

---

### ⚠️ Observaciones importantes

- El denominador:
  
  $$
  2x^{(k-1)} - x^{(k)} - x^{(k-2)}
  $$
  
  no debe ser cercano a cero.

- La estimación de $m^{(k)}$ puede ser inexacta, pero aun así mejora la convergencia.

---

### 🎯 Idea clave

👉 El método **no necesita conocer la multiplicidad**,  
porque la **aprende automáticamente** durante las iteraciones.

---

### 💬 Frase resumen

**“El método adaptativo estima la multiplicidad observando el comportamiento de las iteraciones.”**

---


## 🎯 Ideas clave 

✔ Newton estándar usa derivada en cada paso  
✔ Newton modificado usa la multiplicidad $m$  
✔ Newton adaptativo estima $m$  

---

```python
import math                                      # Libreria matematica

# =========================================================
# EJEMPLO 6.11
# f(x) = (x^2 - 1)^p log(x)
# raiz alpha = 1
# multiplicidad m = p + 1
# =========================================================

def f(x, p):
    return (x**2 - 1)**p * math.log(x)           # Funcion del ejemplo

def df(x, p):
    g = x**2 - 1                                 # Definimos g(x)=x^2-1
    return p*g**(p-1)*2*x*math.log(x) + g**p/x   # Derivada de f(x)


# =========================================================
# 1. NEWTON ESTANDAR
# =========================================================

def newton_estandar(p, x0=0.8, tol=1e-10, max_iter=500):
    x = x0                                       # Valor inicial

    for k in range(max_iter):                    # Ciclo de iteraciones
        fx = f(x, p)                             # Evalua f(x_k)
        dfx = df(x, p)                           # Evalua f'(x_k)

        if abs(dfx) < 1e-14:                     # Evita division por cero
            break

        x_new = x - fx/dfx                       # Newton estandar

        if abs(x_new - x) < tol:                 # Criterio de parada
            return x_new, k+1

        x = x_new                                # Actualiza

    return x, max_iter


# =========================================================
# 2. NEWTON MODIFICADO PARA RAICES MULTIPLES
# x_{k+1} = x_k - m f(x_k)/f'(x_k)
# =========================================================

def newton_modificado(p, x0=0.8, tol=1e-10, max_iter=500):
    x = x0                                       # Valor inicial
    m = p + 1                                    # Multiplicidad exacta

    for k in range(max_iter):                    # Ciclo de iteraciones
        fx = f(x, p)                             # Evalua f(x_k)
        dfx = df(x, p)                           # Evalua f'(x_k)

        if abs(dfx) < 1e-14:                     # Evita division por cero
            break

        x_new = x - m*fx/dfx                     # Newton modificado

        if abs(x_new - x) < tol:                 # Criterio de parada
            return x_new, k+1

        x = x_new                                # Actualiza

    return x, max_iter


# =========================================================
# 3. NEWTON ADAPTATIVO
# x_{k+1} = x_k - m_k f(x_k)/f'(x_k)
# =========================================================

def newton_adaptativo(p, x0=0.8, tol=1e-10, max_iter=500):
    xs = [x0]                                    # Lista para guardar iteraciones

    # Primero hacemos dos pasos de Newton estandar
    for _ in range(2):                           # _ repite esto 2 veces, no me importa el contador
        x = xs[-1]                               # Ultimo valor calculado
        fx = f(x, p)                             # Evalua f(x)
        dfx = df(x, p)                           # Evalua f'(x)

        if abs(dfx) < 1e-14:
            return x, len(xs)-1, None

        xs.append(x - fx/dfx)                    # Paso Newton estandar

    m_estimado = None                            # Guardara la ultima estimacion

    for k in range(2, max_iter):                 # Desde k=2
        xk = xs[k]                               # x^(k)
        xkm1 = xs[k-1]                           # x^(k-1)
        xkm2 = xs[k-2]                           # x^(k-2)

        denominador = 2*xkm1 - xk - xkm2         # Denominador de m^(k)

        if abs(denominador) < 1e-14:             # Evita division por cero
            break

        m_estimado = (xkm1 - xkm2) / denominador # Formula adaptativa para m^(k)

        fx = f(xk, p)                            # Evalua f(x_k)
        dfx = df(xk, p)                          # Evalua f'(x_k)

        if abs(dfx) < 1e-14:
            break

        x_new = xk - m_estimado*fx/dfx           # Newton adaptativo

        xs.append(x_new)                         # Guarda nuevo valor

        if abs(x_new - xk) < tol:                # Criterio de parada
            return x_new, len(xs)-1, m_estimado

    return xs[-1], len(xs)-1, m_estimado


# =========================================================
# 4. EJECUCION DEL EJEMPLO 6.11
# =========================================================

x0 = 0.8                                        # Valor inicial del ejemplo
tol = 1e-10                                     # Tolerancia del ejemplo
valores_p = [2, 4, 6]                           # Valores de p considerados

print("Resultados del Ejemplo 6.11")
print("-------------------------------------------------------------")
print(" p | m=p+1 | Newton estandar | Newton adaptativo | Modificado")
print("-------------------------------------------------------------")

for p in valores_p:
    m = p + 1                                   # Multiplicidad de la raiz

    raiz_std, it_std = newton_estandar(p, x0, tol)
    raiz_mod, it_mod = newton_modificado(p, x0, tol)
    raiz_adap, it_adap, m_est = newton_adaptativo(p, x0, tol)

    print(
    f"{p:2d} | {m:5d} | "
    f"{it_std:15d} ({raiz_std:.6f}) | "
    f"{it_adap:9d} ({raiz_adap:.6f}) | "
    f"{it_mod:10d} ({raiz_mod:.6f})"
)

print("-------------------------------------------------------------")
```

* Newton estándar pierde velocidad cuando la raíz es múltiple.
* Newton modificado recupera convergencia rápida si conoces m.
* Newton adaptativo estima m automáticamente.

## Conclusión conceptual

👉 **Newton estándar:**
- Falla en presencia de raíces múltiples.
- Pierde la convergencia cuadrática.
- Se vuelve muy lento (convergencia lineal).

👉 **Newton modificado (teórico):**
- Está diseñado para corregir el problema de raíces múltiples.
- Debería ser el método más eficiente si la multiplicidad \( m \) es conocida.
- En la práctica puede fallar o volverse inestable por efectos numéricos.

👉 **Newton adaptativo:**
- Funciona sin conocer previamente la multiplicidad \( m \).
- Ajusta dinámicamente el método durante las iteraciones.
- Es más robusto en la práctica frente a problemas numéricos.

La tabla no muestra el valor de la raíz, sino el número de iteraciones necesarias para aproximarla. En el Ejemplo 6.11, la raíz exacta del problema es x = 1, y todos los métodos intentan converger hacia ese valor.

**El 500 indica que el método alcanzó el número máximo de iteraciones sin converger**

## 🔵 ¿Por qué $\log(x)$ aporta multiplicidad 1 en $x=1$?

### 📌 Evaluación en la raíz

$$
\log(1) = 0
$$

Por lo tanto, $x=1$ es una raíz.

---

### 📌 Derivada

$$
\frac{d}{dx}\log(x) = \frac{1}{x}
$$

Evaluando en $x=1$:

$$
\log'(1) = 1 \neq 0
$$

---

### 🎯 Conclusión

Si se cumple que:

- $f(1)=0$
- $f'(1)\neq 0$

Entonces la raíz es **simple (multiplicidad 1)**.

---

### 🔍 Aproximación local

Cerca de $x=1$, la función se comporta como:

$$
\log(x) \approx (x-1)
$$

o de forma equivalente:

$$
\log(x) \sim (x-1)
$$

---

### 💡 Interpretación

- La función tiene un solo factor $(x-1)$
- No hay potencias mayores de $(x-1)$

✔ Por eso aporta **multiplicidad 1**

---

### 💬 Frase clave

**“Como $\log'(1) \neq 0$, la raíz en $x=1$ es simple y aporta multiplicidad 1.”**
